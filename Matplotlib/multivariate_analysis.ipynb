## Multivariate Analysis using Seaborn
- Scatter Plots
- Bar Charts
- Box Plots
- Heatmaps
- Clusters Maps
- Pairplots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
titanic = sns.load_dataset('titanic')
iris = sns.load_dataset('iris')
tips.head()
```

```python
titanic.head()
```

```python
iris.head()
```

---

### Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(x = tips['total_bill'], y = tips['tip'], hue = tips['sex'], size = tips['size'], style = tips['day'])
plt.show()
```

---

### Bar Plot (Numerical - Categorical)

```python
sns.barplot(x = titanic['class'], y = titanic['age'], hue=titanic['sex'])
plt.show()
```

---

### Box Plot (Numerical - Categorical)

```python
sns.boxplot(x = titanic['sex'], y = titanic['age'], hue=titanic['survived'])
plt.show()
```

---

### Distplot (Numerical - Categorical)

📌 `Note` - sns.distplot( ) is deprecated so you have to use either histplot or displot( )

```python
sns.histplot(titanic[titanic['survived']==0]['age'], kde = True, fill = False)
sns.histplot(titanic[titanic['survived']==1]['age'], kde = True, fill = False)
plt.show()
```

---

### HeatMap (Categorical - Categorical)

```python
sns.heatmap(pd.crosstab(titanic['class'],titanic['survived']))
plt.show()
```

```python
(titanic.groupby('Embarked').mean()['Survived']*100) # Analysis
```

---

### ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic['parch'],titanic['survived']))
plt.show()
```

---

### Pairplot

```python
sns.pairplot(iris,hue='species')
plt.show()
```

---

### Lineplot (Numerical - Numerical)

```python
sns.lineplot(x = titanic['age'], y = titanic['age'].value_counts())
plt.show()
```
