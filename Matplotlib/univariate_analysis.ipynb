## Univariate Analysis using Matplotlib & Seaborn
- Bar Plots
- Pie Charts
- Histograms
- Distribution Plots
- Box Plots
- Violin Plots
- Strip Plots
- CDF Plots
- Quantile Plots
- Rug Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

---

### Bar Plots
ðŸ“Œ`Note` - Bar plots are used for count analysis in univariate analysis

```python
plt.figure(figsize = (15, 7))
plt.bar(df['alive'].unique(), df['alive'].value_counts())
plt.show()
```

```python
plt.figure(figsize=(15, 7))
sns.barplot(x = df['who'].unique(), y = df['who'].value_counts())
plt.show()
```

---

### Pie Charts

```python
plt.pie(df['alive'].value_counts(), autopct='%.2f')
plt.show()
```

ðŸ“Œ `Note` - Seabornâ€™s API is designed to focus on statistical applications and provides improved defaults for many matplotlib functionalities, but it does not include pie charts.<br><br>
ðŸ“Œ `Seaborn CatPlots` - Seaborn catplot (kind='count') change bar chart to pie chart

```python
sns.catplot(data = df['sex'], kind='count', height=5)
plt.show()
```

---

### Histograms

```python
plt.figure(figsize = (15, 7))
plt.hist(df['age'], bins=20)
plt.show()
```

```python
plt.figure(figsize = (15, 7))
sns.histplot(df['age'], kde=True)
plt.show()
```

---

### Distribution Plots
ðŸ“Œ `Important` - Density axis is a Kernal Density Function / PDF

```python
sns.displot(df['age'], kind='kde')
plt.show()
```

---

### Box Plots
ðŸ“Œ `Note` - Box plots are generally used to detect the outliers

```python
sns.boxplot(df['age'])
plt.show()
```

ðŸ“Œ `boxenplots` - Used to analysis boxplots of various columns side by side

```python
sns.boxenplot([df['age'], df['fare']])
plt.show()
```

---

### Violin Plots
ðŸ“Œ `Note` - Its a combination of `box plots` and `kde plot`

```python
sns.violinplot(df['age'])
plt.show()
```

ðŸ“Œ `Note` - Though its a multiplevariate Analysis

```python
sns.catplot(x = df['class'], y = df['fare'], kind='violin', hue=df['sex'], split=True)
```

---

### Strip Plots

```python
sns.stripplot(df['age'])
plt.show()
```

---

### Rug Plots

```python
sns.rugplot(df['age'])
plt.show()
```
