> A strip plot is a type of data visualization that displays individual data points along a single axis, often used for visualizing the distribution of a variable across different categories.
> 1. **Structure**: In a strip plot, data points are plotted along one axis (usually the y-axis), with categories represented along the other (often the x-axis). Points can be "jittered" to avoid overlap and show the density of data.<br><br>
> 2. **Usage**: Itâ€™s particularly useful for visualizing small to medium-sized datasets, allowing for quick identification of distribution patterns, clusters, and outliers.<br><br>
> 3. **Comparison**: Strip plots are often used in conjunction with other plots (like box plots) to provide a detailed view of data distribution within each category.<br><br>
> 4. **Customization**: You can adjust point transparency, size, and color to enhance clarity and highlight specific groups or trends.<br><br>
> 5. **Limitations**: While informative, strip plots can become cluttered with larger datasets, making it harder to discern individual points.

```python
sns.catplot(
    x = tips['day'],
    y = tips['total_bill'],
    kind = 'strip',
    hue = tips['sex'],
    height=7,      # height of the plot
    aspect=2.14,    # width/height ratio
    jitter = 0.08
)
plt.show()
```

    <p style="padding: 8px;color:white;"><b><b><span style='color:#2ae4f5'>4 |</span></b> Swarm Plots </b></p>
</div>

> A swarm plot is an enhanced version of a strip plot, designed to display individual data points while avoiding overlap. Here are some important aspects of swarm plots:<br><br>
> 1. **Structure**: Similar to strip plots, swarm plots display data points along one axis (usually the y-axis) for different categories on the other (x-axis). However, swarm plots use a "swarming" algorithm to arrange points in a way that minimizes overlap.<br><br>
> 2. **Distribution Visualization**: Swarm plots effectively show the distribution of data points and can reveal patterns, clusters, and outliers while maintaining the individual data visibility.<br><br>
> 3. **Comparison with Strip Plots**: While both plots show data points, swarm plots provide a clearer representation, especially for larger datasets, because they prevent points from stacking on top of each other.<br><br>
> 4. **Customization**: You can customize colors, shapes, and sizes of the points to convey additional information, such as grouping or categorical variables.<br><br>
> 5. **Use Cases**: Swarm plots are commonly used in fields like biology and social sciences to visualize experimental data, allowing for easy comparisons across different groups.<br><br>
> 6. **Limitations**: Although they improve clarity compared to strip plots, swarm plots can still become crowded with very large datasets, and the arrangement might be less intuitive in some cases.

```python
sns.catplot(
    x = tips['day'],
    y = tips['total_bill'],
    kind = 'swarm',
    hue = tips['sex'],
    height=7,      # height of the plot
    aspect=2.14,    # width/height ratio
)
plt.show()
```

</div>

ðŸ“Œ `Note` - Violin plots combine box plots and density plots to provide a comprehensive view of the distribution of a variable across multiple categories. They display the dataâ€™s probability density on each side, allowing for visual comparisons of the distribution shape, central tendency, and spread within different groups. In multivariate analysis, violin plots are particularly useful for understanding the relationships and variations across multiple dimensions simultaneously.

```python
sns.catplot(
    x = tips['day'],
    y = tips['total_bill'],
    kind = 'violin',
    hue = tips['sex'],
    height = 7,
    aspect = 2.14,
    split = True
)
plt.show()
```
