### DataFrame Creation Methods  
Various techniques to create a Pandas DataFrame, including from dictionaries, lists, CSV files, and NumPy arrays.

```python
# using Series - name considered as index in dataframe
student_data = [
    pd.Series([100,80,10], name = 'Marks'),
    pd.Series([100,80,10], name = 'Sok'),
    pd.Series([100,80,10], name = 'welkfm'),
    pd.Series([100,80,10], name = 'dkf')
]
pd.DataFrame(student_data)
```

```python
# For changing the names of columns
temp = pd.DataFrame(student_data)
temp.columns = ['c1', 'c2', 'c3']
temp.head()
```

```python
# Custom Index
pd.DataFrame(student_data, index=['i1', 'i2', 'i3', 'i4'])
# Length must be equal
```

```python
# Define the Series
a = pd.Series([100, 80, 10], name='Marks')
b = pd.Series([100, 80, 10], name='Sok')
c = pd.Series([100, 80, 10], name='welkfm')

# Create DataFrame with Series as columns
df = pd.DataFrame({'col1': a, 'col2': b, 'col3': c})
df
```

```python
student = [
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90],
]
students = pd.DataFrame(student,columns=['iq','marks','package'])
students.head()
```

```python
# using dicts

student_dict = {
    'name':['nitish','ankit','rupesh','rishabh','amit','ankita'],
    'iq':[100,90,120,80,0,0],
    'marks':[80,70,100,50,0,0],
    'package':[10,7,14,2,0,0]
}

students = pd.DataFrame(student_dict)
students.set_index('name',inplace=True)
students
```

```python
# write a function that can return the track record of 2 teams against each other
def track_record(team1, team2):
  mask = ((ipl['Team1'] == team1) & (ipl['Team2'] == team2) | (ipl['Team1'] == team2) & (ipl['Team2'] == team1))
  # print(mask) # Correct
  new_df = ipl[mask]
  print((new_df['WinningTeam'] == team1).sum())
  print((new_df['WinningTeam'] == team2).sum())

track_record('Chennai Super Kings', 'Mumbai Indians')
```

```python
# Correlation and Covariance
df = pd.DataFrame(
    {
        'a': [1,2,3,4],
        'b': [3.5, 67, 9.0, 5.6],
        'c': [9.5, 6.7, 1.1, 8.9],
        'd': [3.3, 0, 2.3, 9.9]
    }
)

df.corrwith(df, axis = 1)
```

```python
df.corrwith(df, axis = 0)
```

```python
df['b'].corr(df['a'])
```

```python
df['b'].cov(df['a'])
```

```python
df.corr()
# correlation of each column with all the availabel columns in the datafrmae
```
